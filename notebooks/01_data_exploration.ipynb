```python
import pandas as pd

# Load dataset
df = pd.read_csv("../data/raw/results.csv")

# =========================
# Basic Overview
# =========================
print("=" * 50)
print("FIRST 5 ROWS")
print("=" * 50)
print(df.head())

print("\n" + "=" * 50)
print("DATASET INFORMATION")
print("=" * 50)
df.info()

print("\n" + "=" * 50)
print("STATISTICAL SUMMARY")
print("=" * 50)
print(df.describe())

print("\n" + "=" * 50)
print("MISSING VALUES")
print("=" * 50)
print(df.isnull().sum())

print("\n" + "=" * 50)
print("DATASET SHAPE")
print("=" * 50)
print(df.shape)

print("\n" + "=" * 50)
print("COLUMN NAMES")
print("=" * 50)
print(df.columns.tolist())

# =========================
# General Statistics
# =========================
print("\n" + "=" * 50)
print("NUMBER OF MATCHES")
print("=" * 50)
print(len(df))

print("\n" + "=" * 50)
print("NUMBER OF UNIQUE TEAMS")
print("=" * 50)
print(len(set(df["home_team"]) | set(df["away_team"])))

print("\n" + "=" * 50)
print("DATE RANGE")
print("=" * 50)
print(df["date"].min(), "->", df["date"].max())

# =========================
# Tournament Statistics
# =========================
print("\n" + "=" * 50)
print("TOP 15 TOURNAMENTS")
print("=" * 50)
print(df["tournament"].value_counts().head(15))

print("\n" + "=" * 50)
print("TOP 20 HOST COUNTRIES")
print("=" * 50)
print(df["country"].value_counts().head(20))

# =========================
# Goal Statistics
# =========================
print("\n" + "=" * 50)
print("AVERAGE GOALS")
print("=" * 50)
print(f"Home Goals : {df['home_score'].mean():.2f}")
print(f"Away Goals : {df['away_score'].mean():.2f}")

# =========================
# Most Common Scorelines
# =========================
print("\n" + "=" * 50)
print("TOP 20 MOST COMMON SCORELINES")
print("=" * 50)

scorelines = (
    df.groupby(["home_score", "away_score"])
      .size()
      .sort_values(ascending=False)
      .head(20)
)

print(scorelines)
```
